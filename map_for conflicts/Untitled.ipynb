```python
import dash
from dash import html
import base64
from dash.dependencies import Input, Output


app = dash.Dash(__name__) 


# Encode the local sound file.
sound_filename = '/home/adzavon/Downloads/y2mate.com - TOOFAN  CEST PAS NORMAL Clip Officiel.mp3'  # replace with your own .mp3 file
encoded_sound = base64.b64encode(open(sound_filename, 'rb').read())


app.layout = html.Div(children=[
    html.H1(children='Demo for Audio with Dash'),

    html.Div(children='''
        Click the button to play your local .mp3 sounds.
    '''),


    html.Button(id="button1", children="Click me for sound"),
    html.Div(id="placeholder", style={"display": "none"})])


@app.callback(Output("placeholder", "children"),
              [Input("button1", "n_clicks")],
              )
def play(n_clicks):
    if n_clicks is None:
        n_clicks = 0
    if n_clicks != 0:
        return html.Audio(src='data:audio/mpeg;base64,{}'.format(encoded_sound.decode()),
                          controls=False,
                          autoPlay=True,
                          )


if __name__ == '__main__':
    app.run_server(debug=True, port=8051)
```

```python
import dash
from dash import html
from dash.dependencies import Input, Output
import pyttsx3

app = dash.Dash(__name__)
engine = pyttsx3.init()

def speak_text(text):
    engine.say(text)
    engine.runAndWait()

app.layout = html.Div(children=[
    html.H1(children='Welcome to My Web App'),
    html.Button('Say Welcome', id='say-button')
])

@app.callback(Output('say-button', 'n_clicks'),
              [Input('say-button', 'n_clicks')])
def say_welcome(n_clicks):
    if n_clicks is not None:
        speak_text("Welcome to you")

if __name__ == '__main__':
    app.run_server(debug=True)
```
